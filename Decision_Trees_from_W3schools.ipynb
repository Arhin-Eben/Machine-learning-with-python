**DECISION TREES**

```python
import pandas as pd

df = pd.read_csv("/content/drug200.csv")
df
```

```python
#Drop column 'Cholestrol' and 'Drug'
#df = df.drop(['Cholesterol', 'Drug'], axis=1)
#df
```

```python
#To make a decision tree, all data has to be numerical
#We have to convert the non numerical columns 'Sex','BP' and 'Drug' into numerical values
d = {'M':1,'F':0}
df['Sex'] = df['Sex'].map(d)

d = {'LOW':0,'NORMAL':1,'HIGH':2}
df['BP'] = df['BP'].map(d)

d= {'DrugY':0,'drugX':1,'drugA':2,'drugC':3,'drugB':4}
df['Drug'] = df['Drug'].map(d)

d = {'HIGH':2,'NORMAL':1,'LOW':0}
df['Cholesterol'] = df['Cholesterol'].map(d)

df
```

```python
#separate the feature columns from the target column
features = ['Age','Sex','Drug','Cholesterol','Na_to_K']
X = df[features]
y = df['BP']

print(X)
print(y)
```

```python
#Now we can create the actual decision tree, fit it with our details. Start by importing the modules we need:
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)

tree.plot_tree(dtree, feature_names=features)
plt.show()
```

**Predict Values**

```python
print(dtree.predict([[40, 10, 4, 7, 1]]))
```
